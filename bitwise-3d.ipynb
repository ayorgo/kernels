```python
import matplotlib.pyplot as plt
from operator import and_, or_, xor, lshift
from mpl_toolkits.mplot3d import Axes3D
from itertools import product, starmap
```

```python
def plot(xy, z):
    fig = plt.figure(1, figsize=(9, 8))
    ax = Axes3D(fig)
    ax.scatter([t[0] for t in xy], [t[1] for t in xy], z)
```

```python
x = range(100)
y = range(100)
xy = list(product(x, y))
```

```python
# %matplotlib notebook
%matplotlib inline
plot(xy, list(starmap(and_, xy)))
```

```python
plot(xy, list(starmap(or_, xy)))
```

```python
plot(xy, list(starmap(xor, xy)))
```
